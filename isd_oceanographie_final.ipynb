```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv("bottle.csv")
```

```python
data_preprocessed = data.dropna(axis=1, thresh=50000)
```

```python
data_preprocessed = data_preprocessed.drop("Depth_ID", axis=1)
data_preprocessed = data_preprocessed.drop("Sta_ID", axis=1)
```

```python
data_preprocessed = data_preprocessed.fillna(0.0)
```

```python
y = data_preprocessed.pop("T_degC")
```

```python
x = data_preprocessed
```

```python
# Diviser les dataframes x et y en ensembles d'entraînement et de test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Diviser les ensembles d'entraînement en ensembles d'entraînement et de validation
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=42)
```

```python
# Normalisation des caractéristiques
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
print("Taille de l'ensemble d'entraînement :", len(x_train))
print("Taille de l'ensemble de validation :", len(x_val))
print("Taille de l'ensemble de test :", len(x_test))
```

```python
# Créer une instance du modèle
regression_model = LinearRegression()

# Entraîner le modèle
regression_model.fit(x_train, y_train)

# Prédire sur l'ensemble de validation
y_pred = regression_model.predict(x_val)

# Calculer les métriques d'évaluation
mse = mean_squared_error(y_val, y_pred)
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

# Afficher les métriques
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
# Définir le modèle de régression linéaire
regression_model = LinearRegression()

# Effectuer une validation croisée avec 5 plis
scores = cross_val_score(regression_model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')

# Convertir les scores en MSE positifs
mse_scores = -scores

# Afficher les scores de validation croisée
print("Cross-Validation MSE Scores:", mse_scores)

# Afficher la moyenne des scores
print("Mean Cross-Validation MSE:", mse_scores.mean())
```

```python
knn_model = KNeighborsRegressor(n_neighbors=1)

# Entraîner le modèle
knn_model.fit(x_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_KNN = knn_model.predict(x_test)

# Calculer les métriques d'évaluation
mse = mean_squared_error(y_test, y_pred_KNN)
mae = mean_absolute_error(y_test, y_pred_KNN)
r2 = r2_score(y_test, y_pred_KNN)

# Afficher les métriques
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
knn_model = KNeighborsRegressor(n_neighbors=1)

# Effectuer une validation croisée avec 5 plis
scores = cross_val_score(knn_model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')

# Convertir les scores en MSE positifs
mse_scores = -scores

# Afficher les scores de validation croisée
print("Cross-Validation MSE Scores:", mse_scores)

# Afficher la moyenne des scores
print("Mean Cross-Validation MSE:", mse_scores.mean())
```

```python
# Créer une instance du modèle de forêt aléatoire
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Entraîner le modèle
rf_model.fit(x_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred_rf = rf_model.predict(x_test)

# Calculer les métriques d'évaluation
mse = mean_squared_error(y_test, y_pred_rf)
mae = mean_absolute_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

# Afficher les métriques
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
